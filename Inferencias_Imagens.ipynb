```python
from ultralytics import YOLO
import random
import os
import cv2
```

```python
modelo_nano = YOLO('bestn.pt')
modelo_small = YOLO('bests.pt')
```

```python
caminho_pasta = './Eliteserien_Particionado/frames/test'

todas_imagens = os.listdir(caminho_pasta)

imagens_sorteadas = random.sample(todas_imagens, 3)

caminhos_completos = [os.path.join(caminho_pasta, img) for img in imagens_sorteadas]

print("Imagens escolhidas:", caminhos_completos)
```

### Fazendo Inferência nas 3 Imagens Escolhidas

```python
resultadosn = modelo_nano(caminhos_completos)

for resultado in resultadosn:
    resultado.show()
    print("Class ID:", resultado.boxes.cls)
    print("Confidence:", resultado.boxes.conf)
    print("Coordinates:", resultado.boxes.xyxy) 
        
```

```python
caminho_salvar_nano = './Resultados_Individuais/nano'

for resultado, caminho_original in zip(resultadosn, caminhos_completos):

    nome_arquivo = os.path.basename(caminho_original)

    imagem_anotada = resultado.plot()

    cv2.putText(
        imagem_anotada,
        nome_arquivo,
        (20, 50),
        cv2.FONT_HERSHEY_SIMPLEX,
        1,
        (0, 255, 0),
        2
    )

    nome_salvar = os.path.join(caminho_salvar_nano, nome_arquivo)

    cv2.imwrite(nome_salvar, imagem_anotada)

    print(f"Salvo: {nome_salvar}")
```

```python
resultadoss = modelo_small(caminhos_completos)

for resultado in resultadosn:
    resultado.show()
    print("Class ID:", resultado.boxes.cls)
    print("Confidence:", resultado.boxes.conf)
    print("Coordinates:", resultado.boxes.xyxy) 
```

```python
caminho_salvar_nano = './Resultados_Individuais/small'

for resultado, caminho_original in zip(resultadoss, caminhos_completos):

    nome_arquivo = os.path.basename(caminho_original)

    imagem_anotada = resultado.plot()

    cv2.putText(
        imagem_anotada,
        nome_arquivo,
        (20, 50),
        cv2.FONT_HERSHEY_SIMPLEX,
        1,
        (0, 255, 0),
        2
    )

    nome_salvar = os.path.join(caminho_salvar_nano, nome_arquivo)

    cv2.imwrite(nome_salvar, imagem_anotada)

    print(f"Salvo: {nome_salvar}")
```
